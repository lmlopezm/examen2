```python
chrome_options = webdriver.ChromeOptions()
chrome_options.add_argument('--headless')
chrome_options.add_argument('--no-sandbox')
chrome_options.add_argument('--disable-dev-shm-usage')
driver = webdriver.Chrome('chromedriver',chrome_options=chrome_options)
```

```python
url = 'https://es.wikipedia.org/wiki/Anexo:Presidentes_de_Colombia#Lista_de_Presidentes_de_la_República_de_Colombia'
```

```python
driver.get(url)
```

```python
df_periodo = pd.DataFrame(columns=['Periodo','Presidente','Partido politico'])

periodo = driver.find_elements_by_xpath('/html/body/div[3]/div[3]/div[5]/div[1]/table[6]/tbody/tr/td[2]')
presidente = driver.find_elements_by_xpath('/html/body/div[3]/div[3]/div[5]/div[1]/table[6]/tbody/tr/td[4]')
partido = driver.find_elements_by_xpath('/html/body/div[3]/div[3]/div[5]/div[1]/table[6]/tbody/tr/td[5]')

for i in range(len(periodo)):
  df_periodo = df_periodo.append({'Periodo': periodo[i].text, 
                                  'Presidente': presidente[i].text, 
                                  'Partido politico': partido[i].text}, 
                                  ignore_index=True
                                  )
```
